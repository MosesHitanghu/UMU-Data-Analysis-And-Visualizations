<h4> Loading data into data frames </h4>

```python
#Loading titanic data and displaying the first 5 records
titanic_df = pd.read_csv("Titanic-Dataset.csv", sep=",", header=0)
titanic_df.head(5)
```

<h4>  Loading supermarket sales data </h4>

```python
sales_df = pd.read_csv("supermarkets_sales.csv", sep=",", header=0)
sales_df.head(5)
```

<h4>  Loading TB Burden dataset </h4>

```python
tb_burden_df = pd.read_csv("TB_Burden_Country.csv", sep=",", header=0)
tb_burden_df.head(5)
```

<h4> Cleaning datasets using the python reusable class </h4>

```python
#importing the reusable class
from data_cleaner import DataCleaner 
# Creating cleaner instance
cleaner = DataCleaner(strategy="mean", scaling="standard")
```

```python
# store all the dfs in a dictionery
datasets = {
    "TB Burden Data": tb_burden_df,
    "Titanic Data": titanic_df,
    "Sales Data": sales_df
}

```

<h4> Check for missing values </h4>

```python
for name, df in datasets.items():
    print(f"\n===== {name} Missing Values=====")
    cleaner.check_missing_values(df)
```

<h4> Automating Cleaning </h4>

```python

#looping through the dfs in a dictionery

for name, df in datasets.items():
   
    # Apply full cleaning pipeline
    cleaned_df, logs = cleaner.clean(name,df)
    print(f"\n --- {name} Logs ---")
    print("\n".join(logs))
    print("\n--- Cleaned Dataset ---")
    print(cleaned_df)
    
    print(f'\n Checking for missing values in  {name}')
    print('*******************************************')
    cleaner.check_missing_values(cleaned_df)
    
   
```
