<h4> TV Ratings to determine family friendliness; </h4>

<p>From the dataset the following shows are recommended for kids, older, all ages, family and all audiences </P>
<p>Family Friendly Ratings</p>
TV-Y, TV-Y7, TV-Y7-FV, TV-G, G, PG

<p> None Friendly ratings</p>
TV-14, TV-MA, R, PG-13, NC-17, UR


```python
# variables to store content nature
family_friendly = 'Family Friendly'
none_family_friendly = 'Not Family Friendly'
#dictionary for content and their content nature
family_friendly_features = {'TV-Y':family_friendly,'TV-Y7':family_friendly,'TV-Y7-FV':family_friendly,
                            'TV-G':family_friendly,'G':family_friendly,'PG':family_friendly,
                            'TV-14':none_family_friendly, 'TV-MA':none_family_friendly,'R':none_family_friendly,
                            'PG-13':none_family_friendly,'NC-17':none_family_friendly,'UR':none_family_friendly
                            
                           }
```

<h4> Filter out the df to only include the required columns </h4>

```python
important_df = df[important_features]
important_df.head(5)
```

<h4>Creating a new column called comment and map it with values from the family friendly features dictionary</h4>

```python
important_df.loc[:, 'comment'] = important_df['rating'].apply(
    lambda x: family_friendly_features[x] if x in family_friendly_features else "Unknown"
)
important_df
```

<h4> Filtering Actual Family Friendly content </h4>

```python
family_friendy_df = important_df[important_df['comment'] == 'Family Friendly' ]
family_friendy_df
```
